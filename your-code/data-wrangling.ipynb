```python
#primero importamos las librerias a usar

import pandas as pd
```

```python
#previamente desde el sitio descargamos la base de datos como un .csv que ya esta adjuntado para poder usarlo offline
#aqui vamos a abrir nuestra base de datos y transformarla para poder iniciar a trabajar en ella
songs = pd.read_csv('Top_1_000_Songs_To_Hear_Before_You_Die.csv')
```

```python
#transformamos la base de datos en un data frame para poder trabajar mejor con ella

song_df = pd.DataFrame(songs)
```

```python
#se revisa la cantidad de datos nulos de la lista de 994 elementos para decidir que se hara con ella

songnull = song_df.isnull().sum()
songnull[songnull > 0]
```

```python
#eliminamos la columna de SPOTIFY_URL ya que no se como agregar los urls de 22 canciones de manera eficiente

song_df.drop(['SPOTIFY_URL'], axis='columns', inplace=True)
```

```python
#aqui nada mas revisamos que la columna de las 994 canciones y vemos que todos los elementos son correctos

song_df
```

```python
#aqui ya subimos el archivo como un CSV y le cambiamos el nombre para indicar que son 994 y no 1000 canciones

song_df.to_csv('Top_994_Songs_To_Hear_Before_You_Die.csv', index=False)
```

```python
#Aqui decidimos revisar cuantas cnaciones por tema habia 

theme = pd.value_counts(songs["THEME"])
theme
```

```python
#Aqui filtramos las canciones por tema sex (podemos cambiarlos por lo que sea)

songsex = songs[(songs["THEME"] == 'Sex')]
songsex
```

```python
#aqui sacamos los años de la cancion mas vieja y la cancion mas nueva

songsx = songs[['YEAR']].max(axis=0)
songsy = songs[['YEAR']].min(axis=0)
print('La cancion mas antigua es del: ', songsy)
print('La cancion mas actual es del: ', songsx);
```

```python
#para futuras observaciones acomode canciones por año y asi uno podria evitarse el paso de solo pedir maximos y minimos

yearsort = song_df.sort_values(by='YEAR', ascending=True)
yearsort
```
